```python
#   
katok = []
select = -1

#  
def add_data(friend):
    katok.append(None)
    kLen = len(katok)
    katok[kLen - 1] = friend
    
#  
def delete_data(position):
    if position < 0 or position >= len(katok):
        print("   .")
        return 

    kLen = len(katok)
    katok[position] = None

    for i in range(position + 1, kLen):
        katok[i - 1] = katok[i]
        katok[i] = None
    
    del(katok[kLen - 1])

# main
if __name__ == "__main__":
    while (select != 3):
        select = int(input("(1: , 2: , 3: )--> "))
        
        if (select == 1):
            data = input(" --> ")
            add_data(data)
            print(katok)

        elif (select == 2):
            pos = int(input(" --> ")) - 1
            delete_data(pos)
            print(katok)

        elif (select == 3):
            exit

        else:
            print("1~3    .")
            continue
```
