## 사용하는 IDE


```python
# 전역 변수 선언
katok = []
select = -1

# 추가 함수
def add_data(friend):
    katok.append(None)
    kLen = len(katok)
    katok[kLen - 1] = friend
    
# 삭제 함수
def delete_data(position):
    if position < 0 or position >= len(katok):
        print("삭제할 데이터가 존재하지 않습니다.")
        return 

    kLen = len(katok)
    katok[position] = None

    for i in range(position + 1, kLen):
        katok[i - 1] = katok[i]
        katok[i] = None
    
    del(katok[kLen - 1])

# main
if __name__ == "__main__":
    while (select != 3):
        select = int(input("선택하세요(1: 추가, 2: 삭제, 3: 종료)--> "))
        
        if (select == 1):
            data = input("추가할 데이터--> ")
            add_data(data)
            print(katok)

        elif (select == 2):
            pos = int(input("삭제할 순서--> ")) - 1
            delete_data(pos)
            print(katok)

        elif (select == 3):
            exit

        else:
            print("1~3 중 하나의 숫자만 입력하세요.")
            continue
```
